```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from math import pi
from pylab import *
%matplotlib inline

# Manual style for plots
import sys
sys.path.append('../')
import style
```

```python
sys_nm=["pkG","pkA","pkL","pkkg",'pkka',"pkkl"]
dataA=[]
dataB=[]

for i in range(6):
    tmp = pd.read_csv("dih_data/"+str(i)+"_A_"+sys_nm[i]+".csv")
    tmp = tmp.tail(30*400) # last 400fr = last 800ns
    dataA.append(tmp)
    
    tmp = pd.read_csv("dih_data/"+str(i)+"_B_"+sys_nm[i]+".csv")
    tmp = tmp.tail(30*400)
    dataB.append(tmp)
```

```python
def rama_plot(phi_data,psi_data,fr_data,nm,time_min=0,time_max=450,ns_fr=2):
    figure()
    
    fig = scatter(phi_data,psi_data,marker='.',c=fr_data)
    xlim(-180, 180)
    ylim(-180, 180)
    
    title('Dihedral Map:'+nm)
    cbar = colorbar()
    cbar.set_label('Time [ns]',fontsize=style.font_size)
    cbar_ticks=[x * ns_fr for x in range(time_min,time_max,50)]
    cbar.set_ticklabels(cbar_ticks)
    cbar.ax.tick_params(labelsize=style.font_size)
    xlabel(r'$\Phi$ Angle [degrees]',fontsize=style.font_size)
    ylabel(r'$\Psi$ Angle [degrees]',fontsize=style.font_size)
    xticks(fontsize=style.font_size,rotation=45)
    yticks(fontsize=style.font_size)
    
    return fig
```

```python
# Positive peptide
for i in range(6):
    figure()
    
    # Every sixth residue corresponds to DL-X-DL Residue Type:
    tmp = dataA[i][dataA[i]["res"]%6==0]
    tmp = tmp[tmp["phi"]!=0]
    tmp = tmp[tmp["psi"]!=0]
    
    scatter(tmp["phi"], tmp["psi"], marker='.', c=tmp["frame"])
    xlim(-180, 180)
    ylim(-180, 180)
    
    title('Dihedral Map:'+sys_nm[i]+" Lys Containing Chain")
    cbar = colorbar()
    cbar.set_label('Time [ns]',fontsize=style.font_size)
    cbar_ticks=[x * 2 for x in range(0,450,50)]
    cbar.set_ticklabels(cbar_ticks)
    cbar.ax.tick_params(labelsize=style.font_size)
    xlabel(r'$\Phi$ Angle [degrees]',fontsize=style.font_size)
    ylabel(r'$\Psi$ Angle [degrees]',fontsize=style.font_size)
    xticks(fontsize=style.font_size,rotation=45)
    yticks(fontsize=style.font_size)
    #plt.tight_layout()
    #plt.savefig(str(i)+"_DIH_LYS_P.png",dpi=600)
```

```python
# Negative peptide
for i in range(6):
    figure()
    
    tmp = dataB[i][dataB[i]["res"]%6==0]
    tmp = tmp[tmp["phi"]!=0]
    tmp = tmp[tmp["psi"]!=0]

    scatter(tmp["phi"], tmp["psi"], marker='.', c=tmp["frame"])
    xlim(-180, 180)
    ylim(-180, 180)
    
    title('Dihedral Map:'+sys_nm[i]+" GLU Containing Chain")
    cbar = colorbar()
    cbar.set_label('Time [ns]',fontsize=style.font_size)
    cbar_ticks=[x * 2 for x in range(0,450,50)]
    cbar.set_ticklabels(cbar_ticks)
    cbar.ax.tick_params(labelsize=style.font_size)
    xlabel(r'$\Phi$ Angle [degrees]',fontsize=style.font_size)
    ylabel(r'$\Psi$ Angle [degrees]',fontsize=style.font_size)
    xticks(fontsize=style.font_size,rotation=45)
    yticks(fontsize=style.font_size)
    #plt.tight_layout()
    #plt.savefig(str(i)+"_DIH_GLU_P.png",dpi=600)
```
