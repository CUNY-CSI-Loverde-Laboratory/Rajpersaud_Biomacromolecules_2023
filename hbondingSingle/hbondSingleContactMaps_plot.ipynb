```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
import numpy as np
import pyblock
```

```python
peptide_colors = ["#FFE4E1","#CD5C5C","#CD2626","#BFEFFF","#00BFFF","#104E8B"]
sys=["pkG","pkA","pkL","pkkg","pkka","pkkl"]
data=[]

# Positive peptide
for i in sys:
    tmp = pd.read_csv("matrix_data/matrix"+i+"_A.out",header=None)
    data.append(tmp)

#data[0] #pkG
#gen1 = 1100ns in 550 fr
#gen2 = 1000ns in 500fr

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("Intramolecular Hbonding Fraction Lys-Containing Strand (Single Pairs)")
fig.supxlabel("DONER")
fig.supylabel("ACCEPTOR")
fig.tight_layout()

k=0
for i in range(2):
    for j in range(3):
        sns.heatmap(ax=axes[i, j], data=data[k],linewidth=0.5, vmin=0,vmax=0.2,cmap="jet")
        #ax.titleOffsetTrans=100
        k+=1
#fig.suptitle('')
#plt.savefig("pos_single_HBcontacts.png",dpi=600)
```
